```python
import json
import pandas as pd
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
import plotly.io as pio
pio.renderers.default = 'browser'   # displays in a new tab in the default browser
```

```python
kenya_counties_area = json.load(open("G:\My Drive\Jenga school\Python\Labs\Final project\Workspace\kenyan-counties.geojson",'r')) #json file with county codes
kenya_counties = json.load(open("G:\My Drive\Jenga school\Python\Labs\Final project\Workspace\counties.json",'r')) #json file with county co-ordinates
```

```python
# extracts the county code and the county name which can be added to the csv file as a new column
county_id_map = {}
for feature in kenya_counties:
    feature['id']=feature['name']
    county_id_map[feature['code']]=feature['id']
```

```python
county_id_map  # county codes respectively from the couties.json fie
```

```python
#creates a dictionary of the counties names 
county_id_map_area = {}
for feature in kenya_counties_area['features']:
    feature['id']=feature['properties']['COUNTY']
    county_id_map_area[feature['properties']['COUNTY3_ID']] = feature['id']
```

```python
county_id_map_area  # from the geojson file
# make sure the county names below match the county names in the csv file for accurate plotting.
```

```python
df = pd.read_csv("G:\My Drive\Jenga school\Python\Labs\Final project\Workspace\Covid vaccination status as per 8th December 2022.csv")
```

```python
df.head()
```

```python
df['County code'] = county_id_map # adds column county code to csv
```

```python
df['id'] = county_id_map_area  # adds a new column to the csv file
```

```python
#calculates the percentage of adults vaccinated in each county then adds a new column to the csv file
current = df['Number of adults fully vaccinated']
target = df['Target for full vaccination in adults']
percentage = round((current/target)*100,1)
percentage
df['County vaccination status in %']=percentage
```

```python
total_centres =np.sum(df['Number of approved vaccination centres']) # calculates total number of vaccination centres
centres = df['Number of approved vaccination centres']
per_county = round((np.divide(centres,total_centres))*100,1)
df['Vaccination centres as a % of country total'] = per_county
```

```python
df   # calling the dataframe to see what is currently in the csv file
```

```python
# calculating the bottom 5 vaccinated counties
a = df[['County vaccination status in %','County']]
b = a.sort_values(['County vaccination status in %'])
g = lowest_vaccinated_counties = b[:5]
h = lowest_vaccinated_counties['County'].tolist() #convert the data frame to a list so that you convert to string below.
k=','.join(h)
k
```

```python
# calculating the top 5 vaccinated counties
c=df[['County vaccination status in %','County']]
d=c.sort_values(['County vaccination status in %'],ascending=False)
highest_vaccinated_counties = d[:5]
e= highest_vaccinated_counties['County'].tolist() # this is done first because if you convert df direct to string, it includes unnecesary information such as data type
f = ','.join(e)
f
```

```python
# calculating percentage of country that is vaccinated
total_vaccinated =np.sum(df['Number of adults fully vaccinated'])
total_target = np.sum(df['Target for full vaccination in adults'])
total_percent_vaccinated = round((total_vaccinated/total_target)*100,1)
total_percent_vaccinated
```

```python
# calculating counties with the least number of  vaccination centres.
cen = df[['County','Number of approved vaccination centres']]
cen_2 = cen.sort_values(['Number of approved vaccination centres'])
cen_low = cen_2[:5]
cen_low_3 = cen_low['County'].tolist()
cen_low_4=','.join(cen_low_3)
cen_low_4
```

```python
# First choropleth map that shows percentage of adults vaccinated per county
fig = px.choropleth(df, locations='County',
                    geojson = kenya_counties_area, 
                    color = 'County vaccination status in %', 
                   hover_data = ['County code','Number of adults fully vaccinated','Target for full vaccination in adults','Number of approved vaccination centres'])
fig.update_geos(fitbounds="locations",
                    visible=False,   # visibility of base layers
                   framewidth = 5)
fig.update_layout(title_text = 'COVID VACCINATION STATUS AS OF 8TH DECEMBER 2022',
                  title_font_size = 18,
                  title_pad_l = 450,
                  width = 1800,
                  height = 900,
                  clickmode = "select")  #allows a county to be selected and highlighted against the other counties
fig.add_annotation(xref ='x domain',  # puts the annotation with reference to x axis
                   yref = "y domain", # puts the annotation with reference to y axis
                   x = 0.2,
                   y = 0.2,
                   text = "The top 5 vaccinated counties are:",
                   showarrow=False,
                   font=dict(
                      family = "sans serif",
                       color="LightSeaGreen",
                       size = 16))

fig.add_annotation(xref ='x domain',  # puts the annotation with reference to x axis
                   yref = "y domain", # puts the annotation with reference to y axis
                   x = 0.2,
                   y = 0.18,
                   text=f,
                  showarrow=False,
                  font=dict(
                      family = "sans serif",
                      color="LightSeaGreen",
                       size = 16))

fig.add_annotation(xref ='x domain',  # puts the annotation with reference to x axis
                   yref = "y domain", # puts the annotation with reference to y axis
                   x = 0.2,
                   y = 0.15,
                   text="The bottom 5 vaccinated counties are:",
                  showarrow=False,
                  font=dict(
                      family = "sans serif",
                      color="LightSeaGreen",
                       size = 16))

fig.add_annotation(xref ='x domain',  # puts the annotation with reference to x axis
                   yref = "y domain", # puts the annotation with reference to y axis
                   x = 0.2,
                   y = 0.13,
                   text=k,
                  showarrow=False,
                  font=dict(
                      family = "sans serif",
                      color="LightSeaGreen",
                       size = 16))

fig.add_annotation(xref ='x domain',  # puts the annotation with reference to x axis
                   yref = "y domain", # puts the annotation with reference to y axis
                   x = 0.2,
                   y = 0.10,
                   text="AS of 8th Decemmber 36.8% of the adult population had been vaccinated",
                  showarrow=False,
                  font=dict(
                      family = "sans serif",
                      color="LightSeaGreen",
                       size = 16))
fig.show()
                   
pio.write_html(fig, file='index.html', auto_open=True)
```
